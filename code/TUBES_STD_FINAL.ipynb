##**MERGE Dataset**

```python
#races_df & pitStop_df
F1Dataset_merge = pd.merge(races_df, pitStop_df, on='raceId', how='inner')

# Merge with drivers_df
F1Dataset_merge = pd.merge(F1Dataset_merge, drivers_df, on='driverId', how='inner')

# Merge with results_df
F1Dataset_merge = pd.merge(F1Dataset_merge, results_df, on=['raceId', 'driverId'], how='inner')

#circuits_df
F1Dataset_merge = pd.merge(F1Dataset_merge, circuits_df, on='circuitId', how='inner')

#constructors_df
F1Dataset_merge = pd.merge(F1Dataset_merge, constructors_df, on = 'constructorId', how='inner')
F1Dataset_merge.info(); F1Dataset_merge.head(10)
print(len(F1Dataset_merge))
```

##**FILTER**

```python
start_date = '2014'
end_date = '2021'

F1Dataset_filter = F1Dataset_merge[(F1Dataset_merge['year'] >= int(start_date)) & (F1Dataset_merge['year'] <= int(end_date))]
F1Dataset_filter.info()
F1Dataset_filter.head(10)
print(f'Total Barisnya adalah : {len(F1Dataset_filter)}')
```

###**Feature Engineering**

```python
F1Dataset_filter['pit_duration_sec'] = F1Dataset_filter['milliseconds']/1000
```

```python
F1Dataset_filter['pit_duration_sec'] = pd.to_numeric(F1Dataset_filter['pit_duration_sec'], errors='coerce')
```

```python
F1Dataset_filter['date'] = pd.to_datetime(F1Dataset_filter['date'], errors='coerce')
F1Dataset_filter['raceId'] = F1Dataset_filter['raceId'].astype(int)
F1Dataset_filter['driverId'] = F1Dataset_filter['driverId'].astype(int)
```

```python
F1Dataset_filter = F1Dataset_filter.drop(columns=['milliseconds'])
```

```python
total_races_global = results_df.groupby('driverId').size()

F1Dataset_filter['driver_exp'] = F1Dataset_filter['driverId'].map(total_races_global)

def categorize_experience(races):
    if races < 80:
        return 'Rookie'
    elif races <= 200:
        return 'Mid-experience'
    else:
        return 'Veteran'

F1Dataset_filter['experience_category'] = F1Dataset_filter['driver_exp'].apply(categorize_experience)
```

###**Outlier Handling**

```python
F1Dataset_filter.describe()
```

```python
# Remove Outlier dengan IQR
Q1 = F1Dataset_filter['pit_duration_sec'].quantile(0.25)
Q2 = F1Dataset_filter['pit_duration_sec'].quantile(0.50)
Q3 = F1Dataset_filter['pit_duration_sec'].quantile(0.75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

F1Outlier = F1Dataset_filter[(F1Dataset_filter['pit_duration_sec'] < lower_bound) | (F1Dataset_filter['pit_duration_sec'] > upper_bound)]
print(f'Total data dengan outlier : {len(F1Dataset_filter)}')
print(f'Total data outlier : {len(F1Outlier)}')

F1Dataset = F1Dataset_filter[(F1Dataset_filter['pit_duration_sec'] >= lower_bound) & (F1Dataset_filter['pit_duration_sec'] <= upper_bound)]
print(f'Q1 = {Q1}, Q2 = {Q2}, Q3 = {Q3}')
print(f'Total data tanpa outlier : {len(F1Dataset)}')
print("__"*50)


# VISUALISASI
sns.set_theme(style="whitegrid")

#SCATTER PLOT
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Before)]
sns.scatterplot(x='date', y='pit_duration_sec', data=F1Dataset_filter, ax=axes[0], color='salmon')
axes[0].set_title('Scatter Plot: Durasi Pit (Sebelum Remove Outlier)')
axes[0].set_xlabel('Tanggal Balapan')
axes[0].set_ylabel('Durasi Pit Stop (Detik)')

# After
sns.scatterplot(x='date', y='pit_duration_sec', data=F1Dataset, ax=axes[1], color='skyblue')
axes[1].set_title('Scatter Plot: Durasi Pit (Setelah Remove Outlier)')
axes[1].set_xlabel('Tanggal Balapan')
axes[1].set_ylabel('Durasi Pit Stop (Detik)')
plt.tight_layout()
plt.show()

#BOX PLOT
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Before
sns.boxplot(x='pit_duration_sec', data=F1Dataset_filter, ax=axes[0], color='salmon')
axes[0].set_title('Box Plot: Distribusi Durasi Pit (Sebelum Remove Outlier)')
axes[0].set_xlabel('Durasi Pit Stop (Detik)')
axes[0].set_ylabel('Sumbu Distribusi')

# After
sns.boxplot(x='pit_duration_sec', data=F1Dataset, ax=axes[1], color='skyblue')
axes[1].set_title('Box Plot: Distribusi Durasi Pit (Setelah Remove Outlier)')
axes[1].set_xlabel('Durasi Pit Stop (Detik)')
axes[1].set_ylabel('Sumbu Distribusi')
plt.tight_layout()
plt.show()

#HISTOGRAM
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

#Before
sns.histplot(x='pit_duration_sec', data=F1Dataset_filter, kde=True, ax=axes[0], color='salmon')
axes[0].set_title('Histogram: Kepadatan Durasi Pit (Sebelum Remove Outlier)')
axes[0].set_xlabel('Durasi Pit Stop (Detik)')
axes[0].set_ylabel('Jumlah Kejadian (Frekuensi)')

#After
sns.histplot(x='pit_duration_sec', data=F1Dataset, kde=True, ax=axes[1], color='skyblue')
axes[1].set_title('Histogram: Kepadatan Durasi Pit (Setelah Remove Outlier)')
axes[1].set_xlabel('Durasi Pit Stop (Detik)')
axes[1].set_ylabel('Jumlah Kejadian (Frekuensi)')
plt.tight_layout()
plt.show()

#VIOLIN PLOT
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

#Before
sns.violinplot(x='pit_duration_sec', data=F1Dataset_filter, ax=axes[0], color='salmon')
axes[0].set_title('Violin Plot: Densitas & Rentang (Sebelum Remove Outlier)')
axes[0].set_xlabel('Durasi Pit Stop (Detik)')
axes[0].set_ylabel('Kepadatan Distribusi')

#After
sns.violinplot(x='pit_duration_sec', data=F1Dataset, ax=axes[1], color='skyblue')
axes[1].set_title('Violin Plot: Densitas & Rentang (Setelah Remove Outlier)')
axes[1].set_xlabel('Durasi Pit Stop (Detik)')
axes[1].set_ylabel('Kepadatan Distribusi')
plt.tight_layout()
plt.show()

round(F1Dataset.describe())
```

## Ringkasan Alur Data

Proses pengolahan data dari dataset mentah hingga siap dianalisis melalui tahapan berikut:

1. **`F1Dataset_merge`:** Hasil gabungan (*Inner Join*) dari **5 dataset mentah** (`races`, `pit_stops`, `results`, `circuits`, dan `constructors`) berdasarkan ID.
2. **`F1Dataset_filter`:** Subset data hasil filtering **Era Turbo-Hybrid (2014–2021)** yang menghasilkan **5.818 baris**. Pada tahap ini dilakukan *Feature Engineering* pembuatan kolom `pit_duration_sec` ($\text{milliseconds} / 1000$) dan penyesuaian tipe data.
3. **`F1Outlier`:** Variable yang menampung **432 data outlier** dengan durasi pit stop di luar batas IQR .
4. **`F1Dataset` (Final):** Dataset bersih final sebanyak **5.386 baris** setelah melalui pemangkasan (*trimming*) data outlier. Data ini tervalidasi 0 Missing Value dan 0 Duplikat.

#**STD : 1  Pada lap berapa pit stop paling sering dilakukan dan bagaimana pola pit stop sepanjang balapan Formula 1?**

- Mengidentifikasi lap pit stop yang paling sering terjadi selama balapan.
- Menemukan pit rush peroid (periode lonjakan pit stop).



```python
lap_data_series = F1Dataset['lap'].dropna().astype(int)
lap_list = lap_data_series.tolist()
lap_summary_tuple = (min(lap_list), max(lap_list), len(lap_list))
lap_counts = lap_data_series.value_counts()

# Ambil top 5 lap tersibuk, ubah ke bentuk tuple (lap, frekuensi), lalu simpan ke List
top_5_laps = lap_counts.head(5)
pit_rush_list = []
for lap, freq in zip(top_5_laps.index, top_5_laps.values):
    pit_rush_list.append((lap, freq))
```

```python
print("Isi lap_list:")
print(lap_list[:20])

print("\nTuple Ringkasan:")
print(lap_summary_tuple)

print("\nFrekuensi setiap lap:")
print(lap_counts)

print("\nTop 5 Lap Tersibuk:")
print(top_5_laps)

print("\nPit Rush List:")
print(pit_rush_list)
```

```python
# 2. MENAMPILKAN HASIL ANALISIS
print("             ANALISIS FREKUENSI LAP PIT STOP F1")
print(f"Rentang Lap Pit Stop : Lap {lap_summary_tuple[0]} sampai Lap {lap_summary_tuple[1]}")
print(f"Total Transaksi Pit  : {lap_summary_tuple[2]:,} kali")
print(f"Lap Paling Sibuk     : Lap {top_5_laps.index[0]} ({top_5_laps.values[0]:,} pit stop)")

print("        PIT RUSH PERIOD (5 LAP DENGAN FREKUENSI TERTINGGI):")

temp_list = list(pit_rush_list)
rank = 1
while temp_list:
    lap, freq = temp_list.pop(0)
    print(f"Peringkat {rank}: Lap {lap} -> dilakukan sebanyak {freq:,} kali")
    rank += 1
```

```python
# Langkah 1: Menghitung jumlah pit stop yang dilakukan tiap pembalap di setiap balapan
pit_counts_per_driver = F1Dataset.groupby(['circuit_name', 'raceId', 'driverId']).size().reset_index(name='jumlah_pit_stop')

# Langkah 2: Menghitung rata-rata frekuensi pit stop untuk setiap sirkuit
circuit_pit_frequency = pit_counts_per_driver.groupby('circuit_name')['jumlah_pit_stop'].mean().reset_index()
circuit_pit_frequency = circuit_pit_frequency.sort_values(by='jumlah_pit_stop', ascending=False)

# Langkah 3: Menghitung rata-rata durasi pit stop (waktu di pit lane) untuk setiap sirkuit
circuit_pit_duration = F1Dataset.groupby('circuit_name')['pit_duration_sec'].mean().reset_index()
circuit_pit_duration = circuit_pit_duration.sort_values(by='pit_duration_sec', ascending=False)

# DATA KALKULASI
print("="*65)
print("   TOP 10 SIRKUIT DENGAN RATA-RATA FREKUENSI PIT STOP TERTINGGI")
print(circuit_pit_frequency.head(10).to_string(index=False, formatters={'jumlah_pit_stop': '{:,.2f} kali'.format}))

print("\n" + "="*65)
print("   TOP 10 SIRKUIT DENGAN RATA-RATA DURASI PIT STOP TERLAMA")
print(circuit_pit_duration.head(10).to_string(index=False, formatters={'pit_duration_sec': '{:,.2f} detik'.format}))

# VISUALISASI
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(
    x='jumlah_pit_stop',
    y='circuit_name',
    data=circuit_pit_frequency.head(15),
    palette='viridis'
)
plt.title('Top 15 Sirkuit dengan Rata-rata Frekuensi Pit Stop Terbanyak per Pembalap (2014-2021)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Rata-rata Jumlah Pit Stop per Balapan', fontsize=12)
plt.ylabel('Nama Sirkuit', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(
    x='pit_duration_sec',
    y='circuit_name',
    data=circuit_pit_duration.head(15),
    palette='magma'
)
plt.title('Top 15 Sirkuit dengan Rata-rata Durasi Pit Stop Terlama (2014-2021)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Rata-rata Durasi Pit Stop (detik)', fontsize=12)
plt.ylabel('Nama Sirkuit', fontsize=12)
plt.tight_layout()
plt.show()
```

#**STD : 3 Apakah pembalap dengan pengalaman lebih banyak (diukur dari jumlah race yang telah diikuti) cenderung memiliki durasi pit stop yang lebih singkat dibanding pembalap yang lebih sedikit pengalamannya?**

```python
kategori_order = ['Rookie', 'Mid-experience', 'Veteran']

# Langkah 1: Kalkulasi statistik deskriptif (Mean, Median, Count) per kategori
summary_std4 = F1Dataset.groupby('experience_category')['pit_duration_sec'].agg(['mean', 'median', 'count']).reindex(kategori_order)

# Menampilkan hasil analisis numerik
print(" ANALISIS PERBANDINGAN RATA-RATA DURASI PIT STOP PER KATEGORI")
print(summary_std4.to_string(formatters={
    'mean': '{:,.2f} detik'.format,
    'median': '{:,.2f} detik'.format,
    'count': '{:,} kali'.format
}))
# Langkah 2: Visualisasi menggunakan Box Plot (Bebas dari FutureWarning)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.boxplot(
    x='experience_category',
    y='pit_duration_sec',
    hue='experience_category',
    data=F1Dataset,
    palette='Set2',
    order=kategori_order,
    legend=False
)

plt.title('Perbandingan Distribusi Durasi Pit Stop berdasarkan Pengalaman Pembalap (2014-2021)', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Kategori Pengalaman Pembalap', fontsize=11)
plt.ylabel('Durasi Waktu Pit Stop (Detik)', fontsize=11)
plt.tight_layout()

plt.show()
```
